5.5.4 Write a function that multiplies every element of a list by itself, then returns the new list.  Then, write a version that uses a comprehension inside of the function.

```python
def list_squarer(my_list):
    new_list = []
    for item in my_list:
        new_list.append(item**2)
        
    return(new_list)

this_list = [1,2,3,4]
next_list = list_squarer(this_list)
print(next_list)
```

```python
def list_squarer_comp(my_list):
    return [x**2 for x in my_list]

print(list_squarer_comp([1,2,3,4]))
```
