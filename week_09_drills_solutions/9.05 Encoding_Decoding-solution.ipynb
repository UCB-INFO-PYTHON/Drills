9.5.1. What is the unicode representation for the following character: ☕

```python
char = '☕'
encoded_char = char.encode('utf-8')

print(encoded_char)
```

9.5.2. Convert the following unicode characters to ascii: "\u0050\u0079\u0074\u0068\u006f\u006e"

```python
text = "\u0050\u0079\u0074\u0068\u006f\u006e"
print(text.encode('ascii'))
```

9.5.3. Convert the following unicode characters to ascii: "\u0050\u0079\u0074\u0068\u2604\u006e" so that it 'escapes' the character not in the encoding.

```python
# This gives us an error so we need to escape the faulty character
text2 = "\u0050\u0079\u0074\u0068\u2604\u006e"
print(text2.encode('ascii'))
```
