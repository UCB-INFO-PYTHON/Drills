9.6.1. Format the following lists into a nice table, with the float numbers having an accuracy to 2 decimals places. Use the ```%``` method.  The titles are printed first and then the list of items.

````python
table_title = ['word', 'word count', '# of documents', 'average word count per document']
table_list = [['the', 10000, 827, 12.09789842],
              ['a', 9876, 826, 11.95641646],
              ['glue', 1234, 40, 30.85],
              ['Mississippi', 1, 1, 1 ]]
````

````python
print("%-12s  %-10s  %-15s  %-20s" % 
      (table_title[0],
       table_title[1],
       table_title[2],
       table_title[3]))

print('-' * 80)
for item in table_list:
    table = ("%-12s  %10.0f  %12.0f  %20.2f" % 
          (item[0], 
           item[1], 
           item[2], 
           item[3])) 
    print(table)
````

9.6.2. Now use the ```{}``` method to format the list of dictionaries below!

````python
table_title = ['word', 'word count', '# of documents', 'average word count per document']
table_dict = [{'word':'the', 'count':10000, 'doc':827, 'avg':12.09789842},
              {'word':'a', 'count':9876, 'doc':826, 'avg':11.95641646},
              {'word':'glue', 'count':1234, 'doc':40, 'avg':30.85},
              {'word':'Mississippi', 'count':1, 'doc':1, 'avg':1 }]
````

````python
print("{:<12s}  {:<10s}  {:<15s}  {:<20s}".format( 
      table_title[0],
      table_title[1],
      table_title[2],
      table_title[3]))

print('-' * 80)
for item in table_dict:
    table = "{word:<12s} {count:10.0f} {doc:12.0f} {avg:20.2f}".format(
        word=item['word'], 
        count=item['count'],
        doc=item['doc'],
        avg=item['avg']
    )
    print(table)
````
